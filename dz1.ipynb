## Задача 2. 
На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, 
а возвращает общее количество фруктов на складе.

```python
def calculate_total_fruit_count():
    fruit_counts = {}

    n = int(input("Введите количество фруктов: "))

    for i in range(n):
        fruit = input("Введите название фрукта: ")
        count = int(input("Введите количество фрукта: "))

        if fruit in fruit_counts:
            fruit_counts[fruit] += count
        else:
            fruit_counts[fruit] = count

    total_count = sum(fruit_counts.values())

    return total_count


total_fruit_count = calculate_total_fruit_count()
print("Общее количество фруктов:", total_fruit_count)
```

## Задача 3. 
Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. 
Удалите такие значения из списка и посчитайте суммарные затраты
[100, 125, -90, 345, 655, -1, 0, 200].
Используйте list comprehensions

```python
expenses = [100, 125, -90, 345, 655, -1, 0, 200]

# Удаление отрицательных элементов
filtered_expenses = [x for x in expenses if x >= 0]

# Вычсление суммы затрат
total_expenses = sum(filtered_expenses)

total_expenses
```

### Задача 4.2  
Найдите выручку компании в зависимости от месяца
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.
Используйте аннотирование типов.

```python
from typing import List, Dict
from datetime import datetime

def calculate_revenue_by_month(dates: List[str], amounts: List[int]) -> Dict[str, int]:
    # Создаем пустой словарь для хранения выручки по месяцам
    revenue_by_month = {}

    # Проходим по спискам дат и сумм покупок одновременно
    for date, amount in zip(dates, amounts):
        # Преобразуем дату из строки в объект datetime
        date_obj = datetime.strptime(date, '%Y-%m-%d')
        # Извлекаем месяц и год из объекта datetime
        month = date_obj.strftime('%Y-%m')

        # Проверяем, есть ли уже запись для данного месяца в словаре
        if month in revenue_by_month:
            # Если запись уже существует, добавляем текущую сумму к существующей
            revenue_by_month[month] += amount
        else:
            # Если записи еще нет, создаем новую запись с текущей суммой
            revenue_by_month[month] = amount

    # Возвращаем словарь с выручкой по месяцам
    return revenue_by_month

# Пример использования функции с вашими данными

purchase_dates = ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05', '2021-10-09', '2021-09-30', '2021-12-22',
                  '2021-11-29', '2021-12-24', '2021-11-26', '2021-10-27', '2021-12-18', '2021-11-09', '2021-11-23',
                  '2021-09-27', '2021-10-02', '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01', '2021-11-09',
                  '2021-12-06', '2021-12-08', '2021-10-09', '2021-10-31', '2021-09-30', '2021-11-09', '2021-12-13',
                  '2021-10-26', '2021-12-09']
purchase_amounts = [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037, 4274, 6275, 4988, 
                    6930, 2971, 6592, 2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]

revenue_by_month = calculate_revenue_by_month(purchase_dates, purchase_amounts)
print(revenue_by_month)
```
