```python
import pandas as pd

df = pd.read_csv('kc_final.csv')
df.head()
```

### Задача 1.2
Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и 
годом постройки дома.

Если реновации дома не было, то в новом признаке поставьте 0.

```python
df["delta_renovated"]=df["yr_renovated"]-df["yr_built"]
df.loc[df["delta_renovated"]<0,"delta_renovated"]=0

df.head()
```

### Задача 1.3
Создайте признаки года продажи, месяца продажи

```python
df["month_of_sale"]=df["date"].apply(lambda x: x[5:7])
df["year_of_sale"]=df["date"].apply(lambda x: x[:4])
df.head()
```

### Задача 1.4
Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=["date","zipcode","lat","long"],inplace=True)
df.head()
```

## Задача 3

### Задача 3.1

Найдите среднюю стоимость домов в зависимости от количества спален Отсортируйте от меньшей стоимости к большей

```python
bedrooms_df=df.groupby("bedrooms").agg({"price":"mean"}).sort_values("price")
bedrooms_df
```

### Задача 3.2

Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
min_mean_max_price=df.groupby("condition").agg({"price":["min","mean","max"]})
min_mean_max_price
```

### Задача 3.3

Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
view_df=df.groupby(["waterfront","view"]).agg({"house_id":"count"})
view_df
```

### Задача 3.4

Каких домов в зависимости от этажности и количества спален больше?

```python
floors_df=df.groupby(["floors","bedrooms"]).agg({"house_id":"count"}).reset_index().sort_values("house_id",ascending=False).head(1)
floors_df
```

```python
df.pivot_table(index='floors',
               columns='bedrooms',
               values='house_id',
               aggfunc='count',
               fill_value=0)
```

```python
condition_df=df.groupby(["condition","grade"]).agg({"price":"median"}).reset_index()
condition_df
df.pivot_table(index='condition',columns='grade', values='price', aggfunc='median',fill_value=0)
```
